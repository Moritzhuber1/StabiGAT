```python
import pandas as pd
```

```python
data= pd.read_csv('data.csv', index_col=0, sep=';')
```

```python
data
```

```python
temp = pd.to_numeric(data["melting point [°C]"], errors="coerce")

invalid_rows = data[temp.isna()]

print(invalid_rows[["name", "melting point [°C]"]])
```

```python
data["melting point [°C]"] = data["melting point [°C]"].str.replace("<", "", regex=False).str.strip()
```

```python
temp = pd.to_numeric(data["melting point [°C]"], errors="coerce")

invalid_rows = data[temp.isna()]

print(invalid_rows[["name", "melting point [°C]"]])
```

```python
data.to_csv('new_data.csv')
```
